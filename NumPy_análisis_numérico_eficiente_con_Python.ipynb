#Formulas

$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$


La solución para la pendiente es:

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

Y para la intersección:
 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

Y obtenemos la ecuación de la recta:

$$y = \alpha x + \beta $$




```python
n = np.size(Moscu)
y = Moscu
x = fechas
```

$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - (\sum x)^2} $$

```python
a = (n * np.sum(x*y)-np.sum(x)*np.sum(y))/(n*np.sum(x**2) - (np.sum(x)**2))
```

 $$\beta = \frac{\sum y - \alpha \sum x}{n} $$

```python
b = np.mean(y) - a*np.mean(x)
```

```python
print(f'La pendiente: {a}\nLa interseccion: {b}')
```

$$y = \alpha x + \beta $$

```python
y = a *x+b
```
